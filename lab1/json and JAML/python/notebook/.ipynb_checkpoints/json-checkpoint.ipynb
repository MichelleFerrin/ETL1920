## JAML

Creamos una funcion que nos permita cargar y leer y guardar el JAML.

```python
import yaml

def read_yaml_file(filename):
    with open(filename, 'r') as stream:
        try:
            print(yaml.safe_load(stream))
        except yaml.YAMLError as exc:
            print(exc)
```

```python
countries = [{'City': 'New York', 'Country': 'USA'},
         {'City': 'Madrid', 'Country': 'Spain'}]
```

```python
with open('countries.yaml', 'w') as f:
    
    data = yaml.dump(countries, f)
```

```python
read_yaml_file("countries.yaml")
```
